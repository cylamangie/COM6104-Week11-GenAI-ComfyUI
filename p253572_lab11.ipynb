```python
# =========================== Exercise 6 =================================
import json
import requests

# Read the workflow from local disk
with open("exercise5.json", "r", encoding="utf-8") as file:
    workflow = json.load(file)
    # customize the prompt
    # workflow["6"]["inputs"]["text"] = "McDonald,girl,boy,dating"

print(workflow)

# Example condition variable (replace with your own logic)
condition = "B"  # try "A", "B", "C"

# Modify the conditioning node inputs for different poses
if condition == "A":
    workflow["6"]["inputs"]["text"] = "girl standing with arms crossed"
elif condition == "B":
    workflow["6"]["inputs"]["text"] = "girl sitting on a chair"
elif condition == "C":
    workflow["6"]["inputs"]["text"] = "girl running forward"
    
# Submit to ComfyUI (queue)
prompt = {"prompt": workflow}
response = requests.post("http://127.0.0.1:8188/prompt", json=prompt)
print(response.text)
```

```python
## Note the prompt ID!
## It is the information you need to trace your task progress

result = response.json()
pid = result["prompt_id"]
pid
```

```python
response = requests.get(f"http://127.0.0.1:8188/history/{pid}")
history = response.json()
image = None
# Check if an image is generated
if pid in history:
    if "outputs" in history[pid]:
        for x in history[pid]["outputs"]:
            item = history[pid]["outputs"][x]
            if "images" in item:
                image = item["images"]

# The image may not be ready! You may need to run this cell again to check
if image is None:
    print("No image is output (yet)")
else:
    print(image)
```
